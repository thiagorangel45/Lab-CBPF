```python
import ROOT

root_file = ROOT.TFile("MuonLT.root", "READ")

for key in root_file.GetListOfKeys():
    print(key.GetName())

c1 = ROOT.TCanvas("c1", "Canvas for Drawing", 800, 600)
hist2d = root_file.Get("XYBckData")

hist = hist2d.ProfileX()
hist.Rebin(200)

fit = ROOT.TF1("fit", "[0]*exp(-x/[1]) + [2]")
fit.SetParameter(0, 10)
fit.SetParameter(1, 2.5)
fit.SetParameter(2, 1)
hist.Fit(fit)

hist.Draw()
fit.Draw("SAME")
c1.Draw()
```
