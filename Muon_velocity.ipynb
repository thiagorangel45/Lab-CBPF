```python
import ROOT
```

```python
root_file = ROOT.TFile("MuonVelocity.root", "READ")
c1 = ROOT.TCanvas()
for key in root_file.GetListOfKeys():
    print(key)

hist2d = root_file.Get("XYBckData")
hist2d.Draw()
hist = hist2d.ProfileX()
hist.Rebin(100)

hist.Fit("gaus")

hist.Draw()


c1.Draw()
```
